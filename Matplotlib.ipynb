```python
maths=[100,95,92,98]
physics=[93,91,86,82]
chemistry=[96,91,84,76]
l=np.arange(1,5)
width=0.2
plt.xticks(l,['NK','pk','dk','mk'])
plt.bar(l,maths,width,label="maths",color="g",alpha=0.4)
plt.bar(l+width,physics,width,label="Physics",color='b',alpha=0.4)
plt.bar(l-width,chemistry,width,label="chemistry",color="orange",alpha=0.4)
plt.grid(True,color="black")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.legend(bbox_to_anchor=(1,1),loc=2)  #facecolor="red",edgecolor="green",frameon="false"
plt.show()
```

```python
maths=[100,95,92,98]
physics=[93,91,86,82]
chemistry=[96,91,84,76]
l=np.arange(1,5)
width=0.2
plt.xticks(l,['NK','pk','dk','mk'])
m=plt.bar(l,maths,width,label="maths",color="g",alpha=0.4)
p=plt.bar(l+width,physics,width,label="Physics",color='b',alpha=0.4)
c=plt.bar(l-width,chemistry,width,label="chemistry",color="orange",alpha=0.4)

def createlabel(data):
    for i in data:
        height=i.get_height()
        plt.text(i.get_x()+0.1,height+1,height,va="bottom",ha="center")

createlabel(m)
createlabel(p)
createlabel(c)

plt.grid(True)
plt.xlabel("Name")
plt.ylabel("Marks")
plt.legend(bbox_to_anchor=(1,1),loc=2) 
plt.show()
```

```python
#subplot
sub=['maths','phy','chem','bio','comp']
marks1=[99,94,96,91,100]
marks2=[95,89,87,82,79]
marks3=[74,79,84,76,82]

plt.subplot(2,2,1,facecolor="yellow")
plt.plot(sub,marks1)
plt.title("NK")
plt.xlabel("Subjects")
plt.yticks([70,75,80,85,90,95,100])
plt.ylabel("MAkrs")
plt.subplots_adjust(wspace=0.3,hspace=0.5)


plt.subplot(2,2,2)
plt.plot(sub,marks2)
plt.yticks([70,75,80,85,90,95,100],visible=False)
plt.title("DK")
plt.xlabel("Subjects")


plt.subplot(2,2,3)
plt.plot(sub,marks3)
plt.yticks([70,75,80,85,90,95,100])
plt.title("Pk")
plt.xlabel("Subjects")
plt.ylabel("MAkrs")


plt.show()
```

```python
sub=['maths','phy','chem','bio','comp']
marks1=[99,94,96,91,100]
marks2=[95,89,87,82,79]
marks3=[74,79,84,76,82]


plt.subplot2grid(shape=(3,3),loc=(0,0),colspan=2,rowspan=2,facecolor="yellow")
plt.plot(sub,marks1)
plt.title("NK")
plt.xlabel("Subjects")
plt.yticks([70,75,80,85,90,95,100])
plt.ylabel("MAkrs")
plt.subplots_adjust(wspace=0.3,hspace=0.9)



plt.subplot2grid(shape=(3,3),loc=(2,0),colspan=2)
plt.plot(sub,marks2)
plt.yticks([70,75,80,85,90,95,100],visible=False)
plt.title("DK")
plt.xlabel("Subjects")


plt.subplot2grid(shape=(3,3),loc=(0,2),colspan=2,rowspan=2)
plt.plot(sub,marks3)
plt.yticks([70,75,80,85,90,95,100])
plt.title("Pk")
plt.xlabel("Subjects")
plt.ylabel("MAkrs")
plt.tight_layout()

plt.show()
```
