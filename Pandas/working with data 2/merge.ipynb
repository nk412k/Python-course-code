```python
import pandas as pd
import numpy as np
```

```python
df1=pd.DataFrame({'key':['A','A','D','E','D'],'data1':[13,53,65,75,44]})
```

```python
df1
```

```python
df2=pd.DataFrame({'key':['A','T','S'],'data2':[43,64,63]})
```

```python
df2
```

```python
#many to one

pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on='key')
```

```python
pd.merge(df1,df2,on='key',how='left')
```

```python
pd.merge(df1,df2,on='key',how='right')
```

```python
pd.merge(df1,df2,on='key',how='outer')
```

```python
#many to many
df3=pd.DataFrame({'key':['A','A','C','D','D'],'data3':[45,65,23,67,42]})
```

```python
pd.merge(df1,df3,on='key')
```

```python
pd.merge(df1,df3,on='key',how='outer')
```

```python
df4=pd.DataFrame({'key1':['SA','LA','SA'],'key2':[1,1,2],'left_data':[53,54,65]})
```

```python
df4
```

```python
df5=pd.DataFrame({'key1':['SA','NY','LA'],'key2':[1,1,2],'right_data':[23,24,56]})
```

```python
df5
```

```python
pd.merge(df4,df5,how="outer",on='key1')
```

```python
pd.merge(df4,df5,how='outer',on=['key1','key2'])
```

```python
pd.merge(df4,df5,on='key1',how='outer',suffixes=['_left','_right'])
```
